```python
def txt_split(txt):      #split text data in individual text by using txtsplt function
  return txt.split(',')
```

```python
data=df['Products'].apply(txt_split)
data
```

```python
te=TransactionEncoder()
```

```python
df_encoded=te.fit_transform(data)
```

```python
df_encoded
```

```python
te.columns_
```

```python
data=pd.DataFrame(df_encoded,columns=te.columns_)  #creating data frame
```

```python
data.replace([True,False],[1,0],inplace=True)   #converting true false value to 1 and 0
```

lets make data more useful by removing very frequent items and very less frequent items as it will create baises and will dominate the analysis

```python
series=data.sum()
```

```python
series.sort_values()
```

```python
series[(series >= 7500*0.02) & (series <= 7500*0.80)] #removing items with less than 2% and more than 80%
```

```python
new_data = data[series[(series >= 7500*0.02) & (series <= 7500*0.80)].keys().tolist()]
```

```python
new_data
```

```python
# @title ASSOCIATION RULE
```

```python
frequent_itemsets=apriori(new_data,min_support=0.001,use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

**INTERVIEW QUESTIONS**
1-> Lift measures the strength of association between items, calculated as (Confidence/Support of consequent), indicating how much more likely the consequent is when the antecedent occurs.
2-> Support represents the proportion of transactions containing the itemset (e.g., {A, B}), calculated as (Number of transactions with {A, B}) / Total transactions. Confidence measures the probability of the consequent given the antecedent, calculated as (Support of {A, B}) / (Support of A).
3-> Association rules mining faces challenges: handling large datasets, dealing with redundant/unnecessary rules, difficulty in interpreting results, ignoring negative associations, assumption of binary attributes, computational complexity, handling concept drift, and ensuring actionability.
