```python
# Standardization (Z-Score Normalization): Transforms data to mean=0 and standard deviation=1,
# reducing feature dominance and improving model performance, using the formula: z = (x - mu) / sigma.
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    plt.hist(df[col], bins=50)
    plt.title(f'Histogram of {col} (Before Standardization)')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show() # Before standardization: Skewed distributions, varying scales.
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
mu = df[numerical_cols].mean()
sigma = df[numerical_cols].std() #Calculate mean (mu) and standard deviation (sigma) for numerical columns
```

```python
df_standardized = (df[numerical_cols] - mu) / sigma #Standardize numerical columns
```

```python
df_standardized = pd.concat([df_standardized, df.select_dtypes(exclude=['int64', 'float64'])], axis=1)
# Combine standardized numerical columns with original categorical columns
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    plt.hist(df_standardized[col], bins=50)
    plt.title(f'Histogram of {col} (After Standardization)')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show() #-  same scale (mean=0, SD=1).
```
