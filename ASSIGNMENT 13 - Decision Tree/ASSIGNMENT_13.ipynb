```python
# @title 4- decision tree classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
new_df['num'].isnull().sum()
```

```python
new_df['num'].fillna(df['num'].mean(),inplace=True)
```

```python
new_df['num'].isnull().sum()
```

```python
new_df['num'] = (df['num'] > 0).astype(int)

x=new_df.drop('num',axis=1)
y=new_df.num
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
y_train.fillna(y_train.mean(),inplace=True)
y_test.fillna(y_test.mean(),inplace=True)
y_train=y_train.astype(int)
y_test=y_test.astype(int)
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=5)
model.fit(x_train,y_train)
```

```python
#evaluate the model
pred=model.predict(x_test)
```

```python
pd.Series(pred).value_counts()
```

```python
model.feature_importances_
```

```python
np.mean(pred==y_test)   #to measure accuracy
```

```python
pd.crosstab(y_test,pred)
```

```python
accuracy = accuracy_score(y_test,pred)
precision = precision_score(y_test,pred, average='weighted')
recall = recall_score(y_test, pred, average='weighted')
f1 = f1_score(y_test,pred, average='weighted')
roc_auc = roc_auc_score(y_test,pred)

print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1)
print('ROC-AUC:', roc_auc)
```

```python
# @title 5-hyper parameter tuning
```

```python
from sklearn.model_selection import GridSearchCV
#defining grid
grid={
    'max_depth':[2,3,4,5,7,9],
    'min_samples_split':[2,3,4,5,7,10],
    'criterion':['gini','entropy']

}
 # now creating a decision tree classifier
dtc= DecisionTreeClassifier()
  # now grid search
grid_search=GridSearchCV(estimator=dtc,param_grid=grid,cv=5)

  # fiting grid
grid_search.fit(x_train,y_train)

  # getting best parameters
best_param=grid_search.best_params_

  # showing best model
best_model=grid_search.best_estimator_

#evaluating best models
pred=best_model.predict(x_test)
accuracy=accuracy_score(y_test,pred)
precision=precision_score(y_test,pred,average='weighted')
recall=recall_score(y_test,pred,average='weighted')
f1=f1_score(y_test,pred,average='weighted')
roc_auc=roc_auc_score(y_test,pred)

print('best estimators=', best_param)
print('accuracy',accuracy)
print('precision',precision)
print('recall',recall)
print('f1_score',f1)
print('roc_auc_score',roc_auc)
```

**INTERVIEW QUESTIONS**

1-> Decision Trees have several hyperparameters that significantly impact their performance. These include Max Depth, which controls tree complexity and prevents overfitting, Min Samples Split and Min Samples Leaf, which determine the minimum samples required to split a node and be in a leaf node respectively, Max Features, which defines the features considered for splitting, and Criterion, which specifies the splitting criterion such as Gini or Entropy. Adjusting these hyperparameters can prevent overfitting or underfitting, with increased Max Depth and more Max Features potentially leading to overfitting, and higher Min Samples Split and Leaf values potentially resulting in underfitting.

2-> Label Encoding and One-Hot Encoding are two encoding techniques used for categorical data. Label Encoding assigns numerical values to categories, assuming an ordinal relationship, making it suitable for ordinal data. In contrast, One-Hot Encoding creates binary columns for each category, with no assumed ordinal relationship, making it ideal for nominal data. For instance, Label Encoding would represent colors as Red=0, Green=1, and Blue=2, while One-Hot Encoding would represent them as Red=[1,0,0], Green=[0,1,0], and Blue=[0,0,1]. Choosing the correct encoding depends on data characteristics and model requirements.
